```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_size = 2
num_classes = 3
```

```python
class RedeCompetitiva:
  
  def __init__(self, input_size, num_classes):
    self.pesos = np.random.rand(input_size, num_classes)

  def normalizar_pesos(self):
    normas = np.linalg.norm(self.pesos, axis=0)
    self.pesos = self.pesos / normas

  def treinar(self, inputs):
    for x in inputs:
      ativacao = np.dot(self.pesos.T, x)
      vencedor = np.argmax(ativacao)
      self.pesos[:, vencedor] += x
      self.normalizar_pesos()
```

```python
inputs = np.array([[0.5, 0.8], [0.9, 0.3], [0.2, 0.4], [0.6, 0.9]])
rede = RedeCompetitiva(input_size = 2, num_classes = 3)
rede.treinar(inputs)
```

```python
clientes = np.array([[0.1, 0.2], [0.4, 0.5], [0.8, 0.9], [0.3, 0.7], [0.6, 0.4]])
rede.treinar(clientes)

# Assumindo que rede.pesos tem shape (num_centroides, 2)
# 1. Corrija o range do loop para iterar sobre os centroides (shape[0])
for i in range(rede.pesos.shape[0]):
    # 2. Corrija a indexação para pegar (linha i, coluna 0) e (linha i, coluna 1)
    plt.scatter(rede.pesos[i, 0], rede.pesos[i, 1], label=f'Centroid {i+1}')

plt.xlabel('Frequência de Compras')
plt.ylabel('Valor médio das Compras')
plt.title('Centroides da Rede Competitiva')
plt.legend()
plt.show()
```
