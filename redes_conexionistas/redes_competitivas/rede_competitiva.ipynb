```python
import numpy as np
import matplotlib as plt

input_size=2
num_classes=3
```

```python
class RedeCompetitiva:
  
  def __init__(self, input_size, num_classes):
    self.pesos = np.random.rand(input_size, num_classes)

  def normalizar_pesos(self):
    normas = np.linalg.norm(self.pesos, axis=0)
    self.pesos = self.pesos / normas

  def treinar(self, inputs):
    for x in inputs:
      ativacao = np.dot(self.pesos.T, x)
      vencedor = np.argmax(ativacao)
      self.pesos[:, vencedor] += x
```
